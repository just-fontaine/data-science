### Criando colunas numéricas

```python
import pandas as pd

dados = pd.read_csv(
    r'https://raw.githubusercontent.com/alura-cursos/pandas-conhecendo-a-biblioteca/main/base-de-dados/aluguel.csv',
    delimiter=';'
)
```

```python
dados.head()
```

```python
dados['Valor_por_mes'] = dados['Valor'] + dados['Condominio']
```

```python
dados
```

```python
dados['Valor_por_ano'] = dados['Valor_por_mes'] * 12 + dados['IPTU']
dados.head()
```

### Criando colunas categóricas

```python
dados['Descricao'] = dados['Tipo'] + ' em ' + dados['Bairro'] + ' com ' + \
                                        dados['Quartos'].astype(str) + ' quarto(s) ' + \
                                        ' e ' + dados['Vagas'].astype(str) + ' vaga(s) de garagem.'
```

```python
dados.head()
```

```python
dados['Possui_suite'] = dados['Suites'].apply(lambda numero_suites: 'Sim' if numero_suites > 0 else 'Não')
```
