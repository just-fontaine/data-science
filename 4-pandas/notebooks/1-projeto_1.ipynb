## Qual o valor médio de aluguel por tipo de imóvel?

## Removendo os imóveis comerciais

```python
imoveis_comerciais = [
    'Conjunto Comercial/Sala',
    'Prédio Inteiro',
    'Loja/Salão',
    'Galpão/Depósito/Armazém',
    'Casa Comercial',
    'Terreno Padrão',
    'Box/Garagem',
    'Chácara',
    'Loteamento/Condomínio',
    'Sítio',
    'Pousada/Chalé',
    'Hotel',
    'Indústria'
]
```

```python
dados.query(
    '@imoveis_comerciais not in Tipo'
)
```

```python
df = dados.query(
    '@imoveis_comerciais not in Tipo'
)
```

```python
df
```

```python
df['Tipo'].unique()
```

```python
df_preco_tipo = df.groupby('Tipo')[['Valor']].mean().sort_values(by='Valor')

df_preco_tipo.plot(
    kind='barh',
    figsize=(14, 10),
    color='purple'
)
```

## Filtros

### **1. Apartamentos que possuem `1 quarto` e `aluguel menor que 1200`**



```python
df.query(
    'Quartos == 1 and Valor < 1200'
)
```

```python
selecao_1 = df['Quartos'] == 1
selecao_2 = df['Valor'] < 1200
selecao_final = (selecao_1) & (selecao_2)

df[selecao_final]
```

### **2. `Apartamentos` que possuem pelo menos `2 quartos`, `aluguel menor que 3000` e `area maior que 70`**

```python
df.query('Quartos >= 2 and Valor < 3000 and Area > 70').describe().round(2)
```

```python
selecao = (df['Quartos'] >= 2) & (df['Valor'] < 3000) & (df['Area'] > 70)
```

```python
df[selecao]
```
