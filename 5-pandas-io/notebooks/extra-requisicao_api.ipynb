#### Importação das bibliotecas

```python
import requests, json
import pandas as pd
```

```python
# URL para requisiçao

url = r'https://jsonplaceholder.typicode.com/users'
```

```python
# Requisição

request = requests.get(url)
```

```python
resultado = json.loads(request.text)
```

```python
# Visualização do resultado

dados = pd.DataFrame(resultado)

dados.head()
```

```python
# Normalização da coluna de endereço

pd.json_normalize(dados['address'])
```

```python
# Normalização da coluna das empresa

pd.json_normalize(dados['company'])
```
