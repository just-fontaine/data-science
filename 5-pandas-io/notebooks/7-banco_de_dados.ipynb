#### Importação das bibliotecas

```python
import sqlalchemy
from sqlalchemy import create_engine, MetaData, Table, inspect
import pandas as pd
```

#### Criação da engine

```python
engine = create_engine(
    'sqlite:///:memory:'
)
```

```python
caminho = r'C:\Estudos\alura\data-science\5-pandas-io\arquivos\clientes_banco.csv'
```

#### Leitura dos dados

```python
dados = pd.read_csv(
    caminho,
    delimiter=','
)
```

```python
dados.head()
```

#### Transformação para banco de dados

```python
dados.to_sql(
    'clientes',
    con=engine,
    index=False
)
```

```python
inspector = inspect(
    engine
)
```

```python
print(inspector.get_table_names())
```

```python
query = "SELECT * FROM clientes WHERE Categoria_de_renda = 'Empregado'"
```

```python
empregados = pd.read_sql(
    query,
    con=engine
)
```

```python
empregados.head()
```

```python
empregados.to_sql(
    'empregados',
    con=engine,
    index=False
)
```
