### Importação das bibliotecas

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import os
```

```python
sns.set_theme()
```

```python
df = pd.read_csv(
    r'C:\Estudos\alura\data-science\9-dataviz\dados\imigrantes_canada.csv'
)

df = df.set_index('País')
```

```python
top_10 = df.sort_values(by='Total', ascending=False).head(10)
```

```python
# top_10
```

```python
palette = sns.color_palette('dark', 10)
```

```python
sns.barplot(data=top_10, x=top_10.index, y='Total',
            hue=palette, legend=False);
```

```python
sns.barplot(data=top_10,
            x='Total', y=top_10.index,
            hue=palette, legend=False,
            orient='h');
```

```python
ax = sns.barplot(data=top_10,
                 x='Total', y=top_10.index,
                 hue=palette, legend=False,
                 orient='h')

ax.set(title='Países com maior imigração para o Canadá\n(1980-2013)',
       xlabel='Número de imigrantes', ylabel='');
```

```python
def gerar_grafico_paleta(palette):
    fig, ax = plt.subplots(figsize=(8,4))
    ax = sns.barplot(data=top_10, x='Total', y=top_10.index,
                     orient='h', hue=top_10.index, palette=palette, legend=False)
    ax.set_title('Países com maior imigração para o Canadá\n(1980-2013)',
                 fontsize=16)
    ax.set_xlabel('Número de imigrantes', fontsize=13)
    ax.set_ylabel('')

    plt.show()
```

```python
gerar_grafico_paleta('Blues_r')
```

```python
gerar_grafico_paleta('rocket')
```

```python
gerar_grafico_paleta('coolwarm_r')
```

```python
gerar_grafico_paleta('tab10')
```

```python
sns.set_theme(style='dark')
gerar_grafico_paleta('tab10')
```

```python
sns.set_theme(style='whitegrid')
gerar_grafico_paleta('tab10')
```

```python
sns.set_theme(style='white')
gerar_grafico_paleta('tab10')
```

```python
sns.set_theme(style='ticks')
gerar_grafico_paleta('tab10')
```

```python
fig, ax = plt.subplots(figsize=(8,4))
ax = sns.barplot(data=top_10, x='Total', y=top_10.index,
                    orient='h', hue=top_10.index, palette='tab10', legend=False)
ax.set_title('Países com maior imigração para o Canadá\n(1980-2013)',
                fontsize=16)
ax.set_xlabel('Número de imigrantes', fontsize=13)
ax.set_ylabel('')
sns.despine()
```
